# Quantitative Momentum Strategy 

## Objective: 

The purpose of this project is to create a portfolio of 5 high momentum stocks using the quantitative momentum strategy. The dataset is composed of 20 stocks from the automobile and steel industries. 

## Step 1: Importing required libraries

```python
import numpy as np 
import pandas as pd 
import requests  
import math 
from scipy import stats 
```

## Step 2: Importing stock data

```python
# Specify the file path
file_path = "/Users/mazin/Desktop/data/Stock data.csv"

# Load the CSV file into a Pandas DataFrame
data = pd.read_csv(file_path)

# Display the first and last few rows of the DataFrame
data
```

```python
data.drop(columns=['Date'], inplace=True)
data
```

## 3: Calculating the One-Year Price Return

Creating a dataframe with the variables we need

```python
# Create 'final_dataframe' with columns in the specified order
final_dataframe = pd.DataFrame(columns=['Ticker', 'Start Price', 'Ending Price', 'One-Year Price Return', 'Number of Shares to Buy'])

# Populate 'Ticker' column
final_dataframe['Ticker'] = data.columns

# Populate 'Start Price' and 'Ending Price' columns
final_dataframe['Start Price'] = data.iloc[0].values
final_dataframe['Ending Price'] = data.iloc[-1].values

# Display the 'final_dataframe'
final_dataframe
```

```python
# Calculate one-year price return and store it in 'One-Year Price Return' column
final_dataframe['One-Year Price Return'] = ((final_dataframe['Ending Price']-final_dataframe['Start Price']) / final_dataframe['Start Price'])*100 
final_dataframe
```

```python
# Drop the 'Start Price' column
final_dataframe.drop(columns=['Start Price'], inplace=True)

# Rename the 'Ending Price' column to 'Price'
final_dataframe.rename(columns={'Ending Price': 'Price'}, inplace=True)
final_dataframe
```

## 4: Removing low momentum stocks 

The investment strategy that we're building seeks to identify the 5 highest-momentum stocks in the dataset. We will remove all the stocks in our DataFrame that fall below this momentum threshold. We'll sort the DataFrame by the stocks' one-year price return, and drop all stocks outside the top 5.

```python
final_dataframe.sort_values('One-Year Price Return', ascending = False, inplace = True)
final_dataframe = final_dataframe[:5]
final_dataframe.reset_index(drop = True, inplace = True)
final_dataframe
```
